```python
from geopy.distance import geodesic

# Coordonnées de Paris (représentant la France)
france = (48.8566, 2.3522)

# Fonction pour calculer la distance
def calculer_distance(row):
    pays = (row['lat'], row['lon'])
    return geodesic(france, pays).kilometers

# Appliquez la fonction à chaque ligne du DataFrame
distance_pays['distance_france'] = distance_pays.apply(calculer_distance, axis=1)

# Affichez les résultats
distance_final=distance_pays[['iso3','pays', 'distance_france']]
display(distance_final.head(5))
```

```python
# 1. Compter les occurrences de chaque pays
dup_counts = distance_final['pays'].value_counts()

# 2. Filtrer uniquement les pays ayant des doublons (>1 occurrence)
duplicates = dup_counts[dup_counts > 1]

# 3. Convertir en DataFrame explicite
df_duplicates = duplicates.reset_index()
df_duplicates.columns = ['pays', 'nombre_doublons']

# Afficher le résultat
print(df_duplicates.sort_values('pays'))



```

```python
# Groupement par iso3 et calcul de la moyenne
distance_groupby = distance_final.groupby('iso3', as_index=False).agg({
    'pays': 'first',  # Garder le premier nom de pays associé à l'iso3
    'distance_france': 'mean'
})

# Réorganiser les colonnes
df_distance_pays = distance_groupby[['iso3', 'pays', 'distance_france']]

# Afficher le résultat
print(df_distance_pays)

```

```python
# 1. Compter les occurrences de chaque pays
dup_counts = df_distance_pays['pays'].value_counts()

# 2. Filtrer uniquement les pays ayant des doublons (>1 occurrence)
duplicates = dup_counts[dup_counts > 1]

# 3. Convertir en DataFrame explicite
df_duplicates = duplicates.reset_index()
df_duplicates.columns = ['pays', 'nombre_doublons']

# Afficher le résultat
print(df_duplicates.sort_values('pays'))
```
